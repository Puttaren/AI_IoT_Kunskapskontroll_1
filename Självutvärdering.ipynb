# Självutvärdering
## 1. Har något varit utmanande i kursen/kunskapskontrollerna? Om ja, hur har du hanterat det?

Det har varit vissa utmaningar, främst att faktisk komma igång med teorin, men när jag väl satte mig och fokuserade gick det ganska bra med hjälp av boken, en utmärkt PDF om design thinking och några varv av inspelade föreläsningar. De mer praktiska uppgifterna gick så klart inte som en dans, men när jag körde fast tog jag hjälp av AI och mina klasskamrater. 

## 2. Vilket betyg anser du att du ska ha och varför?

VG. Jag anser mig ha uppnått kunskapskraven och målen för kursen, och dessutom engagerat mig utöver bara en G-nivå.   
## 3. Vad har varit roligast i kursen?

Det har varit en allmänt rolig kurs, men de praktiska uppgifterna är alltid roligast eftersom det blir "hands on" och man får sätta tänderna i något "verkligt". 

## 4. Något du vill lyfta fram till Terese?

Jag saknade IoT-delen tidigare i kursen, men fick så jag tålde på slutet. Det hade varit utmärkt med en liten genomgång av vilka AI-modeller som finns tillgängliga, vad de är bra för och vilken man ska välja i vilken situation. Det här är kanske inget som ingår i kursbeskrivningen, men hade passat bra med tanke på kursens namn.
